```python
# Import all required Libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="ticks", color_codes=True)
```

<a id=item2><font size=4 color="229C75">Exploring DataSet</font></a><br>
Let's download the Dataset and analyze

```python
df = pd.read_csv('winequality-red.csv', sep=';')
df.rename(columns = lambda x: x.lower().replace(' ','_'), inplace=True)
df.head()
```

```python
df.describe()
```

### Relationship Visualization

```python
sns.pairplot(df, x_vars=df.columns[0:10], y_vars=df.quality)
```

```python
plt.figure(figsize=(15,6))

plt.subplot(121)
sns.scatterplot(data=df, x='alcohol', y='quality', hue='quality')
plt.title('Alcohol against Quality')

plt.subplot(122)
sns.scatterplot(data=df, x='fixed_acidity', y='quality', hue='quality')
plt.title('Fixed Acidity against Quality')
plt.suptitle('Relationships against Qualit', fontsize=15)
```
