## Numerical Data Plotting
Let's take a data set available with Seaborn itself. We will build a relations plot, which is a kind of FacetGrid.<br>
A **FacetGrid** maps a dataset onto multiple axes arrayed in a grid of rows and columns that corrospond to levels of variables in the dataset.

<li> relplot()
<li> scatterplot()
<li> lineplot()

```python
tips = sns.load_dataset('tips')
tips.tail()
```

```python
dir(sns.FacetGrid)
```

### Plot a Relative graph of "total_bill" and "tip" using "hue","style" and "size"

```python
# FacetGrid
g = sns.relplot(data=tips, x='total_bill', y='tip', hue=tips.smoker, style='time' )
g
```

```python
g = sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", size="size")
# g = sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", size="size", sizes=(15, 200))
g
```

Let's plot the charts alternate to each other

```python
g = sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', col='time')
```

```python
g = sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker', col='size', col_wrap=3, height=3)
```

```python
fmri = sns.load_dataset('fmri')
fmri.tail()
```

```python
g = sns.relplot(data=fmri, x='timepoint', y='signal'
                    , kind='line' # Plot a line
                    , ci='sd' # Take "Standard Deviation" for Confidence Interval (CI)
               )
```

```python
g = sns.relplot(data=fmri, x='timepoint', y='signal'
                            , kind='line' # Use line to change to Line form
                            , ci=False # Disable/Enable Continuous interval
                            , hue='event'
                            )
g
```

```python
g = sns.relplot(data=fmri, x='timepoint', y='signal', hue='region'
                    , style='event'
                    , kind='line'
                    , markers=True # Show markers
                    , dashes=False # Disable dashes
                    , ci=68 # Disable dashes
                    )
```

```python
g = sns.relplot(x = 'timepoint', y = 'signal', hue = 'subject', col = 'region', row = 'event', height=3, kind = 'line', estimator = None, data = fmri)
g
```

### Default plotiing of catplot is a scatterplot

```python
sns.catplot(data=tips, y='total_bill', x='day')
```

```python
sns.catplot(data=tips, y='total_bill', x='day'
            , jitter=False # Align scatter into a single line
           )
```

```python
sns.catplot(data=tips
            , y='total_bill'
            , x='day'
            , hue='sex'
            , kind='swarm'
            , order=['Sun', 'Sat', 'Fri', 'Thur']  # To make your own order
           )
```

```python
sns.catplot(data=tips
            , y='total_bill'
            , x='day'
            , kind='box'
            , hue='time')
```

```python
sns.catplot(data=tips
            , y='total_bill'
            , x='day'
            , kind='boxen')
```

```python
sns.catplot(data=tips
            , y='total_bill'
            , x='day'
            , kind='violin'
            , hue='sex'
            , split=True)
```

Plotting two plots together

```python
g = sns.catplot(data=tips, y='total_bill', x='day', kind='violin', inner=None)
sns.swarmplot(data=tips, y='total_bill', x='day', size=3, color='k', ax=g.ax)
```

## Visualizing Distribution of Data
1. distplot()
2. kdeplot()
3. jointplot()
4. rugplot()

```python
from numpy.random import randn
x = randn(100)
sns.distplot(x)
```

```python
sns.distplot(x
             , kde=True
             , rug=True # Show rugs at the bottom
             , bins=30 # No. of bins
             , hist=False # Hide histogram
            )
```

```python
sns.kdeplot(x, shade=True, cbar=True, bw=1, cut=0.8)
```

```python
x = tips['tip']
y = tips['total_bill']
```

```python
sns.jointplot(x,y)
```

```python
sns.jointplot(x,y, kind='kde', color='m')
```

## Linear Relationship
1. regplot()

```python
sns.regplot(data=tips, x='tip', y='total_bill')
```

```python
sns.lmplot(x = 'size', y = 'tip', data = tips, x_jitter = 0.05)
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', col='day', col_wrap=2)
```

## Controlling Aesthetics

```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1,7):
        plt.plot(x, np.sin(x+i*0.5)*(7-i)*flip)
```

```python
sinplot(-1)
```

```python
# Change the Grid look and feel
sns.set_style('whitegrid')
sinplot(-1)
```

```python
# Change the Grid look and feel
sns.set_style('ticks') # Remove Grids
sinplot(-1)
sns.despine() # Remove the Spines at top and right corner
```

```python
# Change the Grid look and feel
sns.set_style('white')
sinplot(-1)
sns.despine(left=True)
```

```python
# We can control other elements of the plot by passing their values
sns.axes_style()
```

```python
# Change the Grid look and feel
sns.set_style('ticks'
              ,{
                  'xtick.direction': 'in', # Keep ticks direction inside
                  'ytick.right': True, # Show right tick
                  'axes.grid': True # Show Grid
               } 
             )
sinplot(-1)
```

```python
sns.set_context('talk', font_scale=.1) # This affects things like the size of the labels
sinplot()
```
