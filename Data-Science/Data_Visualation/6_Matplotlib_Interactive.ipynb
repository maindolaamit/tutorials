MouseEvent provides use access to what axes we are in, the mouse button that was clicked, any modified keys the user was holding down and hte location of the mouse. 

````python
class LineMaker:
    def __init__(self, ln):
        # Stash the Line2D, 
        self.ln = ln
        self.button_cid = ln.figure.canvas.mpl_connect('button_press_event',
                                                       self.on_button)
        
    def on_button(self, event):
        print(f'button: {event.button!r} @ ({event.xdata}, {event.ydata} + key: {event.key})')
````

````python
fig, ax = plt.subplots()
ln, = ax.plot([1], [1], 'x')
line_marker = LineMaker(ln)
````
