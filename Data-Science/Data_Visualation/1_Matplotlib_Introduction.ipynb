## Introduction
In this Notebook I will be showing the Visualization using matplotlib package.<br>
[Matplotlib](https://matplotlib.org/index.html) is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack. It was introduced by John Hunter in the year 2002.

One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram etc.

A *matplotlib visualization* is a figure onto which is attached one or more axes. Each axes has a horizontal (x) axis and vertical (y) axis, and the data is encoded using color and glyphs such as markers (for example circles) or lines or polygons (called patches). The figure below annotates these parts of a visualization and was created by Nicolas P. Rougier using matplotlib.
![](anatomy.png)

**Let's plot a simple line graph using range() function. x-axis and y-axis will be created automatically to fit the data**<br>
The plot() function is very customizable, various parameters to plot lines, markers, marker type of various size, color and legend.<br>

Let's create a Vector in Numpy and plot the same.<br>

```python
y = array([5,4,5,4])
plot(y)
```

Let's provide the x-axis values to the plot

```python
x = array([8,9,10,14])
plot(x,y)
```

##### Plot an equation 
Let's plot a function where Y is function of X. This way we cen plot the equations.<br>
We can use *title()* method to give the figure a Title.

$$y = x^2 -15 $$

```python
x = array(range(5))
plot(x,x**2-15)
title('An Equation')
```

$$c = \sqrt{a^2 + b^2}$$

```python
a = array(range(10))
b = array([i+5 for i in range(10)])
plot(a,b,sqrt(a**2 + b**2))
title('Square Function')
```

## Linewidth and Markersize
We can control the linewidth and market size by passing *linewidth*(lw) and *markersize*(ms) arguments to plot.<br>
They accept the floating point value and default is 1.

```python
# This graph will have an increased Linewidth and Markersize
plot(y,'x--',lw=2.5,ms=15)
```

```python
# This graph will have an improved Markersize
x=arange(5,10)
plot(x,sin(x),'+--',ms=40)
```
