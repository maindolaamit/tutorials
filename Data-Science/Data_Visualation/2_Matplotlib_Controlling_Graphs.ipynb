## Title, Labels Grid & Lines
The Funciton **title(str)** sets *str* as a title for the graph and gets displayed above the plot area. Below are the argument list<br>

<table align='left'>
    <tr><th>Argument</th><th>Description</th><th>Values</th></tr>
    <tr align='center'><td>fontsize</td><td>Controls the font size</td><td>'large','medium','small',<br> or an acutal size i.e. 50</td></tr>
    <tr align='left'><td>va</td><td>Controls the vertical alignment</td><td>'top','baseline',<br>'bottom','center'</td></tr>
    <tr align='left'><td>ha</td><td>Controls the horizontal alignment</td><td>'center','left','right'</td></tr>

````python
plot(range(5))
title('Testing Title',fontsize=20,va='baseline', ha='right')
````

The function **xlabel()** and **ylabel()** are used to set the x-axis and y-axis labels.<br>
*Both of them accpet the text arguments from the above table*

````python
x = arange(1,3*pi,0.1)
y = sin(x)
plot(x,y,'mo-',lw='.5',)
# stem(x,y,'r')
title('Sine Wave',fontsize=20,va='baseline', ha='center')
xlabel('Degree',fontsize=15)
ylabel('Sin(x)',fontsize=20)
````

Another useful function is **grid()** which is composed of Horizontal and Vertical dashed lines.<br>
We can toggle the grid by calling function **grid(True)** or **grid(False)**.<br>

[**axhline()**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.axhline.html) and [**axyline()**](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.axvline.html) are two other functions which adds a Horizontal line and a Vertical line across the axis<br>

Let's create another Sinwave graph

````python
time = arange(0,10,0.1)
plot(time,sin(time),'.-')
title('Sinwave',fontsize=15,va='baseline')
ylabel('Amplitude = sin(Time)',fontsize=15,color='g')
xlabel('Time',fontsize='15',color='b')
axhline(color='g',y=0)
grid(True)
show()
````

## Ticks & Legends
**Ticks** are the values to show specific points on the cordinates axis which can be either a number or a string.<br>
When we plot a graph the axis adjust and take the default ticks which are nice in most of the scenarios but can be customized.<br>

To control the ticks, use the functions **xticks()** and  **yticks()**. The function take an array holding the tick values as number
and an optional tuple containing text labels. If tuple of lables are not provided, the tick number are used as labels.

To uderstand the ticks better let's draw a graph comparision

````python
R = 1.2
I = arange(0, 4*pi, 0.01)
plot(sin(I)*R, cos(I*.5)*R)
axhline(color='gray')
axvline(color='gray')
grid(True)
title('Ticks as labeles')
xticks([-1,0,1],('Negative','Neutral','Positive'))
yticks(arange(-1.5, 2.0, 1))

# Create a new Figure without Ticks
figure()
plot(sin(I)*R, cos(I*.5)*R)
axhline(color='gray')
axvline(color='gray')
grid(True)
title('Default Ticks')

show()
````

Let's now see another text function **legend()**. The **legend()** function adds a legend box which associates a plot with Text.<br>
In general when you name a graph/plot it is a legend. Let's take the below example

````python
I = arange(0,2*pi,0.1)
plot(sin(I),'go-',cos(I),'rx:')
legend(['Sin','Cos'])
````

The above graph can also be replicated by specifying **label** argument with the **plot()** function call, and then call **legend()** 1with no parameters.

````python
plot(I,sin(I),'go-',label='Sin')
plot(I,cos(I),'rx:',label='cos')
legend()
yticks(arange(-1,1.5,0.5))
xlabel('Time(seconds)')
````

We can control the location of legend box with **loc** parameter to avoid overlapping of legend box with graph line.<br>
Below are the list of values taken by **loc** parameter. Instead of using strings, once can use numbers as well<br>

## Text Rendering
The **text(x,y,str)** function renders the string str on the graph at the coordinates units x,y.<br>
We can modify the text alignment using the arguments passed to Horizontal and Vertical alignment.<br>
The fucntion accepts a lot of other parameters use **help(text)** to view the complete list of arguments.

````python
plot(arange(1,5,1))
text(0,0,'Origin', va='center',ha='center')
````
