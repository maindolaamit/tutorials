## Subplots

**subplot()** fucntion splits the figure into subplots and select the current subplot. Subplots are numbered left to right, top to bottom thus making upper left subplot as 1. Lower right subplot is equivalent to the number of subplots.

Let's compare the sin and cos functions using Subplots. Here we will be creating a Plot with 2 subplots as row and 1 Column.

```python
I = arange(0, 4*pi, 0.1)

subplot(2,1,1)
plot(I,sin(I))
title('Sin Graph')

subplot(2,1,2)
title('Cos Graph')
plot(I,cos(I))
```

Let's use **add_subplots()** to add the subplots to a figure. Here we will create a Figure of 2 by 2 (Total 4 ) Subplots.<br>
We've used the hspace and wspace arguments of subplots_adjust, which specify the spacing along the height and width of the figure, in units of the subplot size (in this case, the space is 40% of the subplot width and height).

```python
fig = figure()

fig.add_subplot(221)
fig.add_subplot(222)
fig.add_subplot(223)
fig.add_subplot(224)

# Put space between Sub-Plots
fig.subplots_adjust(hspace=0.4, wspace=0.4)
```

Let's create the Subplots in Loop and assign some text to them

```python
fig = figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i in range(1,5):
    fig.add_subplot(2,2,i)
    text(0.5, 0.5, str((2,2,i)), ha='center')

        
```

```python
x = linspace(0, 10, 200)
y = sin(x)
fig, ax = subplots()
ax.plot(x, y, 'b-', linewidth=2)
show()
```

Let's create a large grid of subplots, by hide the x- and y-axis labels on the inner plots. For this purpose, **plt.subplots()** is the easier tool to use (note the s at the end of subplots).<br>
Rather than creating a single subplot, this function creates a full grid of subplots in a single line, returning them in a NumPy array.<br>
The arguments are the number of rows and number of columns, along with optional keywords sharex and sharey, which allow you to specify the relationships between different axes<br>
<br>
Here we'll create a 23 grid of subplots, where all axes in the same row share their y-axis scale, and all axes in the same column share their x-axis scale:

```python
ax = subplots(2, 3, sharex='col', sharey='row')
```

Let's draw another Subplots

```python
I = arange(0, 1, 0.01)

# Sub-Plot 1
ax1 = subplot(2,1,1)
plot(I,sin(2*pi*I),'g')
title('Sin Graph', fontsize=15)
ylabel('1 Hz')

# Sub-Plot 2
ax2 = subplot(2,1,2)
plot(I,cos(2*pi*I),'g--')
title('Cos Graph', fontsize=15)
ylabel('1 Hz')

# Add color to the X-Ticklabels
for label in ax1.get_xticklabels():
    label.set_color('r')

for label in ax2.get_xticklabels():
    label.set_color('b')

# Give space between the two plots    
subplots_adjust(hspace=0.6, wspace=0.4)
```

## Axis
**axis()** fucntion controls the behavior of hte x-axis and y-axis, we use it to set new axis ranges by specifying new values : axis([xmin, xmax, ymin, ymax])<br>
If you wish to set only the x-axis or y-axis values we can use **xlim(xmin, xmax)** or **ylim(ymin, ymax)** fucntions.<br>
Apart from range limits the function also accepts the following values : *auto, equal, tight, scaled and off.*<br>

Let's see the below charts by plotting a circle, we can produce a circle by combining Matematical sin and cos functions

```python
R = 1.2
P = arange(0, pi*2, 0.01)
axis('equal')
plot(sin(P)*R,cos(P)*R,'b')
```

Let's create the subplots to examine the circles better. The *subplot()* funciton splits the figure into subplots.<br>
The subplots are numbered from left to right, top to bottom thus upper left subplot is 1 and lower right subplot equal to number of subplots.<br>
I will explain in detail about the subplots 

```python
figure()

subplot(1,2,1)
plot(sin(P)*R,cos(P)*R,'b')
axis('equal')
title('Equal')

subplot(1,2,2)
plot(sin(P)*R,cos(P)*R,'r')
axis('auto')
title('Auto')

figure()
subplot(1,2,1)
plot(sin(P)*R,cos(P)*R)
title('Tight')
axis('tight')

subplot(1,2,2)
plot(sin(P)*R,cos(P)*R)
title('Scaled')
axis('scaled')

show()
```
