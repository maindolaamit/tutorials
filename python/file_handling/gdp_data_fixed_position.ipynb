We can see the data is a fixed format file, with below columns:<br>
    Rank      : Rank of Country based on GDP, width 1:7
    Country   : Name of the Country, width 8:59
    GDP Value : GDP Value of the Country having comma and Dollar symbols, width 60:

```python
print("Rank\tCountry\t{:>50}".format("GDP in Billions"))
print('='*80)
for row in open('gdp_data.txt').readlines():
    rank, country, gdp_value = row[0:7].strip(), row[7:58].strip(), row[58:].strip()
    # Remove $ and convert the GDP to billion
    gdp_value = re.sub('\$|\,','',gdp_value)
    gdp_value = float(gdp_value)/1e9
        
    print("{:<3}\t{:<40}\t{}".format(rank, country, gdp_value))
        
        
```

Let's define a function which will return the N number of Countries and their GDP amount inthe form ot Lists

```python
def get_gdp_data(N=10):
    """A function to read the GDP data from file for first N countries """
    # Let's define the empty list
    countries, gdp = [], []
    # Loop for the data in File
    for i,row in enumerate(open('gdp_data.txt').readlines()):
        rank,country,gdp_value = row[0:7].strip(), row[7:58].strip(), row[58:].strip()
#         print(rank, country, gdp_value)
        # Remove $ and convert the GDP to billion
        gdp_value = re.sub('\$|\,','',gdp_value)
        gdp_value = float(gdp_value)/1e9
        
        
        # Append the data
        gdp.append(gdp_value)
        countries.append(country)
        
        # Exit if number of requests are met
        if(i >= N):
            return (gdp, countries)
        
```

```python
N = 5
gdp, labels = get_gdp_data(N)

print(gdp,labels)
```
